```python
def per(lst):
    
    if len(lst) == 1:
        return [lst]
    p = []
    for i in lst:
        rem = [x for x in lst if x != i]
        l = per(rem)
        
        for k in l:
            p.append([i]+k)
    return p
    
lis = [1,2,3]
p = per(lis)
print(p)
```

```python
def com(lst,n):
    
    if n == 0:
        return [[]]
    c = []
    
    for i in range(len(lst)):
        m = lst[i]
        rem = lst[i+1:]
        k = com(rem,n-1)
        
        for j in k:
            c.append([m] + j)
    return c

print(com([1,2,3],2))
```
